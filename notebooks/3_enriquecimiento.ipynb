## Proyectemos unas variables contra otras

```python
programa_inversion = df_inversiones['PROGRAMA'].value_counts().to_dict()
```

```python
programa_inversion
```

```python
list(programa_inversion.keys())[0:5]
```

```python
df_inversiones['PROGRAMA_INVERSIÓN'] = df_inversiones['PROGRAMA'].apply(lambda x: programa_inversion[x])
```

```python
top_servicios = df_inversiones.groupby('SERVICIO')['INVERSIÓN (MILES DE $ AÑO 2021)'].sum().sort_values(ascending=False).index[0:5]
```

```python
service_mean_inversion = df_inversiones.groupby('SERVICIO')['INVERSIÓN (MILES DE $ AÑO 2021)'].mean()
```

```python
df_inversiones['INVERSION_SERVICIO_PROMEDIO'] = df_inversiones['SERVICIO'].apply(lambda x: service_mean_inversion[x])
```

```python
region_mean_inversion = df_inversiones.groupby('REGIÓN')['INVERSIÓN (MILES DE $ AÑO 2021)'].mean()
```

```python
region_mean_inversion.to_dict()
```

```python
df_inversiones['INVERSION_REGION_PROMEDIO'] = df_inversiones['REGIÓN'].apply(lambda x: region_mean_inversion[x])
```

```python
year_mean_inversion = df_inversiones.groupby('AÑO')['INVERSIÓN (MILES DE $ AÑO 2021)'].mean()
```

```python
year_mean_inversion.to_dict()
```

```python
df_inversiones['INVERSION_AÑO_PROMEDIO'] = df_inversiones['AÑO'].apply(lambda x: year_mean_inversion[x])
```

## Traernos información externa, enriquecer los datos usando una tool de páginas de linkdn

```python
has_word = {}
word = 'conservacion'
for i, ele in df_inversiones.fillna(True).iterrows():
  has_word[ele['NOMBRE']] = True if word in ele['NOMBRE'].lower() else False
```

```python
df_inversiones['HAS_WORD_CONSERVACION'] = df_inversiones['NOMBRE'].apply(lambda x: has_word[x])
```
